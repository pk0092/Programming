# 초물 육안 검사 기록 프로그램

```python
import time
year=time.localtime().tm_year#연
month=time.localtime().tm_mon#월
day=time.localtime().tm_mday#일
hour=time.localtime().tm_hour#시
minute=time.localtime().tm_min#분
second=time.localtime().tm_sec#초
dday=time.localtime().tm_wday#요일(월요일부터 0 ~ 6)
ddday=time.localtime().tm_yday#연 시작에서 몇 일
print(year, month, day, hour, minute, second, dday, ddday)
```
