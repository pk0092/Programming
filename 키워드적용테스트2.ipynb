# 문제부분-문제현상

```python
import win32com.client as wc
excel = wc.Dispatch("Excel.Application")
ex1=excel.Workbooks.Open(r'C:\Users\uia97275\Documents\Programming\pulse\smd1_1.xlsx')
ws1= ex1.ActiveSheet
machine=ws1.Cells(1,1).Value
PAT=[]
CAS=[]

for i in range(4,30):
    part=ws1.Cells(i,5).Value
    if part==None: break
    PAT=PAT+[part]

for j in range(4,30):
    caus=ws1.Cells(j,6).Value
    if caus==None: break
    CAS=CAS+[caus]
    
for p,ppart in enumerate(PAT):
    for c,ccaus in enumerate(CAS):
        order=len(CAS)*p+c+3
        ws1.Cells(order,13).Value=machine
        ws1.Cells(order,14).Value=ppart
        ws1.Cells(order,15).Value=ccaus
ws1.Cells(order+1,13).Value=machine
ws1.Cells(order+1,14).Value="수기입력"
ws1.Cells(order+1,15).Value="수기입력"

ex1.Save()
excel.Quit()
print("Done")
```
