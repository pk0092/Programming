```python
##이게 진짜 @@

import win32com.client as wc #윈도우 리소스 컨트롤 모듈

excel = wc.Dispatch("Excel.Application") #엑셀 접근


#엑셀파일 오픈 & 시트 활성화
ex1=excel.Workbooks.Open(r'C:\Users\uidk3482\Documents\project\적용해보자.xlsx')
ex2=excel.Workbooks.Open(r'C:\Users\uidk3482\Documents\project\키워드_조치자.xlsx')
ws1= ex1.ActiveSheet
ws2= ex2.ActiveSheet


#엑셀파일의 셀 내용을 받을 리스트 변수(리스트화 하는 이유->동작이 20배 더 빠름)
causes=[]
results=[]


#리스트 길이
listlength=1001 ##변경해야할 변수: 조치원인 길이+1


#셀내용들 리스트화
for i in range(1,listlength):
    causes=causes + [ws1.Cells(i,1).Value]
    
    #if ws1.Cells(i,3).Value is not None:
    #    results=results + [ws1.Cells(i,3).Value]
    #else:
    #    results=results + ['?']

    
#2차원으로 분류되있는 키워드들 리스트화(1차원)
colnum=213 ##변경해야할 변수: 키워드 column 개수 + 1
rowlists=[] #하위 키워드 개수 리스트
keywords=[] #1차원으로 된 리스트
for col in range(1,colnum):
    for row in range(1,listlength):
        if type(ws2.Cells(row,col).Value) is not str:
            rowlists=rowlists+[row-1]
            break
        else:
            keywords=keywords+[ws2.Cells(row,col).Value]

print(len(keywords))#참고만


#진짜 키워드들만 Labeling 하기위한 리스트            
rows=[0]#0을 넣은 이유: 처음에 아래 range(row,rows[rnum+1]+1) 부분에서 0부터 탐색하게하려고
temp=0
for rowlist in rowlists:
    if rowlist is not 0:
        temp=temp+rowlist
        rows=rows+[temp]
print(len(rows)) #=column


#키워드번호만 추출
extractnum=[]#그냥 키워드 번호
for cause in causes:
    for knum, keyword in enumerate(keywords):
        if keyword in cause:
            extractnum=extractnum+[knum]
            break
        else:
            extractnum=extractnum+[0]
print(len(extractnum)) #=column-1


#키워드 적용(Labeling)
for num, keynum in enumerate(extractnum):
    for rnum, row in enumerate(rows):
        try:
            if keynum in range(row,rows[rnum+1]+1):
                ws1.Cells(num+1,5).Value=keywords[row]##변경해야할 변수: 컬럼
                break
        except:
            pass
            ws1.Cells(num+1,5).Value='?'##변경해야할 변수: 컬럼
            
#저장 및 종료            
ex1.Save()
excel.Quit()
print("Done")
```
